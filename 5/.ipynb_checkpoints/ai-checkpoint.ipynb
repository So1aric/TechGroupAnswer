```python
import torch
from torch import nn
from torch.optim import Adam
import random
import matplotlib.pyplot as plt
```

```python
class Approx(nn.Module):
    def __init__(self):
        super().__init__()
        self.network = nn.Linear(1, 1)

    def forward(self, x):
        return self.network(x)
```

```python
a = random.uniform(-10, 10)
b = random.uniform(-10, 10)
dataset_x = torch.Tensor([i / 100 for i in range(-100, 100)]).reshape(-1, 1)
dataset_y = torch.Tensor([a * x + b + random.uniform(-1, 1) for x in dataset_x])
```

```python
%matplotlib widget

plt.figure()
plt.scatter(dataset_x, dataset_y)
```

```python
network = Approx()
loss_fn = nn.MSELoss()
optimizer = Adam(network.parameters())
epoch = 30

for e in range(epoch):
    print(f'epoch {e}:')

    output = network(dataset_x)
    loss = loss_fn(dataset_y, output)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f'    loss: {loss.item()}')
    
```
